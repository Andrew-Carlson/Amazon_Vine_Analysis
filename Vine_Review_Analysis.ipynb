```python
# import dependencies
import pandas as pd
```

```python
# read in csv file containing vine table
df = pd.read_csv("./data/vine_table.csv", sep = ",")
df.head(10)
```

```python
df.describe()
```

```python
# filter dataframe by retrieving all the rows where the total_votes count is equal to or greater than 20
total_votes_df = df[df["total_votes"] > 20]
total_votes_df.head(10)
```

```python
total_votes_df.describe()
```

```python
# retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%
helpful_votes_df = total_votes_df[(total_votes_df["helpful_votes"]/total_votes_df["total_votes"]) >= .5]
helpful_votes_df.head(10)
```

```python
helpful_votes_df.describe()
```

```python
# create a new DataFrame or table using helpful_votes_df that retrieves all the rows 
# where a review was written as part of the Vine program (paid), vine == 'Y'
vine_program_df = helpful_votes_df[helpful_votes_df["vine"]=="Y"]
vine_program_df.head(10)
```

```python
vine_program_df.describe()
```

```python
# retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'
not_vine_program_df = helpful_votes_df[helpful_votes_df["vine"]=="N"]
not_vine_program_df.head(10)
```

```python
not_vine_program_df.describe()
```

```python
# Determine the total number of reviews, the number of 5-star reviews, 
# and the percentage of 5-star reviews for the two types of review (paid vs unpaid)
```

```python
# total number of reviews
total_review_count = len(helpful_votes_df)
total_review_count
```

```python
# total number of 5-star reviews
five_star_review_count = (helpful_votes_df["star_rating"] == 5.0).sum()
five_star_review_count
```

```python
# total number of 5-star reviews (paid)
five_star_total_paid = (vine_program_df["star_rating"]== 5.0).sum()
five_star_total_paid
```

```python
# total number of 5-star reviews (unpaid)
five_star_total_unpaid = (not_vine_program_df["star_rating"]== 5.0).sum()
five_star_total_unpaid
```

```python
# percentage of 5-star reviews for the two types of review (paid vs unpaid)
five_star_percent_paid = f'{(five_star_total_paid/five_star_total_unpaid) * 100} %'
five_star_percent_paid
```

```python
info =[{"Total Reviews":total_review_count, 
               "Total 5-Star Reviews": five_star_review_count,
               "Total Vine Program 5-Stars": five_star_total_paid,
               "Total Non-Vine Program 5-Stars": five_star_total_unpaid,
               "Percent 5-Star Vine vs. Non-Vine": five_star_percent_paid}] 
summary_df = pd.DataFrame(info)
summary_df
```
